# **Pregunta 1. Â¿CuÃ¡l es la tendencia de actos delictivos en MÃ©xico a traves de los aÃ±os?**

Para responder la primera pregunta, se obtiene el total de delitos cometidos en un aÃ±o en todas las entidades del paÃ­s. Este total se obtiene para los datos desde el 2015-2020 esperando encontrar alguna tendencia en los datos

# 2015

```python
df_T2015 = df_2015.copy() #Se crea una copia de los datos que se van a trabajar

#Se crea una lista, con el nombre de las columnas de los datos
suma_lista = list(df_T2015)
#Se eliminan de la lista la columnas que no se van a utilizar
suma_lista.remove('Clave_Ent')
suma_lista.remove('AÃ±o')
suma_lista.remove('Cve. Municipio')

#Se agrega la columna "Total", la cual contiene la suma de todos los delitos realizados en un aÃ±o por cada municipio 
df_T2015['Total'] = df_T2015[suma_lista].sum(axis=1)

#Se suman todos los delitos realizados en un aÃ±o para conocer la cifra total
df_T2015 = df_T2015['Total'].groupby(df_T2015['AÃ±o']).sum()
df_T2015 = df_T2015.to_frame().reset_index()
df_T2015
```

# **Pregunta 2. Â¿La violencia tiene una correlaciÃ³n con el PIB que genera cada estado?**

Para responder esta pregunta se tuvo que tomar otro Dataset, esta vez de la pÃ¡gina del INEGI, el cual contiene los datos de PIB por estado desde el aÃ±o 2003 y hasta el aÃ±o 2020. Se escogio analizar el aÃ±o 2019 ya que marca el fin de una tendencia que estaban siguiendo los datos. Ademas el aÃ±o 2020 marco el inicio de la pandemia de covid 19, por lo que para analizar estos datos se tendrian que analizar probablemente mÃ¡s factores

```python
#df_PIB.head(10)
```

Procesando los datos para obtener el PIB por entidad federativa en el aÃ±o 2019

```python
df_PIB1 = df_PIB.copy() #Se hace una copia del Dataset original
df_PIB1 = df_PIB1.drop(['2003','2004','2005','2006','2007',
          '2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2020 P'],axis=1)#Se eliminan las columnas de datos que no se utilizaran
df_PIB1 = df_PIB1.drop([0,1,2,3,4,5], axis=0)#Se eliminan las filas de datos que no se utilizaran
df_PIB1 = df_PIB1.reset_index()
df_PIB1= df_PIB1.drop(df_PIB1[df_PIB1['index']>37].index)#MÃ¡s filas que no se utilizaran
df_PIB1 = df_PIB1.drop(['index'],axis=1)

df_PIB1 = df_PIB1.rename(columns = {'Concepto': 'Entidad', '2019 R':'PIB', 'Total':'Total de delitos cometidos'})#Se renombran las columnas 
#df_PIB1#.head(32)
```

Procesando los datos para obtener el total de delitos cometidos por entidad federativa en el aÃ±o 2019

```python
df_2019C = df_2019.copy()#Se hace una copia del Dataset original

#Se remueven las columnas con valores numericos y que no se utilizaran en nuestra siguiente operaciÃ³n
suma_lista = list(df_2019C)
suma_lista.remove('Clave_Ent')
suma_lista.remove('AÃ±o')
suma_lista.remove('Cve. Municipio')

#Se suman todos los datos cometidos a lo largo del aÃ±o
df_2019C['Total de delitos cometidos'] = df_2019C[suma_lista].sum(axis=1)
#df_2019C.head()

#Se agrupan los datos por entidad federativa
df_2019CF = df_2019C['Total de delitos cometidos'].groupby(df_2019C['Entidad']).sum()
df_2019CF = df_2019CF.to_frame().reset_index()

#df_2019CF
```

Se unen los datos de delitos cometidos por entidad federativa y el PIB por entidad federativa para crear un nuevo Dataset que contenga estos datos

```python
#Se unen los datos del PIB por entidad federativa y los datos de los delitos cometidos por entidad federativa
df_PIBM = pd.merge(df_PIB1,df_2019CF)
df_PIBM = df_PIBM.reindex(columns =['Total de delitos cometidos','PIB','Entidad'])
df_PIBM2 = df_PIBM.copy() 
df_PIBM = df_PIBM.pivot_table('PIB','Total de delitos cometidos','Entidad')
```

# VisualizaciÃ³n de resultados

```python
sns.color_palette('dark')
plt.figure(figsize=(30,10))
cx = sns.heatmap(df_PIBM, cmap='viridis',square = True, robust = True, linewidth = 1, xticklabels = True)
cx.set_title('CorrelaciÃ³n de delitos cometidos y PIB por entidad federativa', fontsize = 20)
cx.set_ylabel('Total de delitos cometidos',fontsize = 20)
cx.set_xlabel('Entidad',fontsize = 20)
plt.show()
```

Se escogÃ­o otra visualizaciÃ³n para poder ver mejor la correlaciÃ³n de los datos PIB - Delitos cometidos

```python
fig = px.scatter(df_PIBM2, x = 'PIB', y = 'Total de delitos cometidos',
    color = 'Entidad', size = 'PIB')
fig.update_layout(title_text = 'RelaciÃ³n de PIB - Delitos cometidos por entidad federativa en MÃ©xico 2019', title_x = 0.42)
fig.show()
```

## Conclusiones



1.   Hay una correlaciÃ³n entre los delitos cometidos por entidad y el PIB
2.   En cuanto mas grande sea el PIB hay mas delitos cometidos
